```python
for i in titanic.columns:
    print(i,':','\n',titanic[i].unique())
```

```python
for i in ['Name','Ticket','Cabin']:
    titanic.drop(i,axis=1,inplace=True)
```

```python
titanic
```

```python
from sklearn.preprocessing import LabelEncoder

lr = LabelEncoder()
```

```python
for i in ['Pclass','Sex','Survived','Embarked']:
    titanic[i] = lr.fit_transform(titanic[i])
```

```python
titanic.shape
```

```python
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic.fillna(value=29,axis=1,inplace=True)
```

```python
plt.figure(figsize=(15,10))
cor = titanic.corr()
sns.heatmap(cor,annot=True)
plt.show()
```

```python
#ip/op

ip = titanic.drop('Survived',axis=1)
op = titanic.Survived
```

```python
#train_test_split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2,train_size=0.8)
```

```python
#Standard Scaler

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x_test = sc.fit_transform(x_test)
x_train = sc.fit_transform(x_train)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()
lg.fit(x_train,y_train)
```

```python
ypred = lg.predict(x_test)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score,accuracy_score

a = accuracy_score(y_test,ypred)
b = recall_score(y_test,ypred)
print(f"accuracy_score : {a} \n recall_score : {b}")
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm = confusion_matrix(ypred,y_test)
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```

```python
#Naive Bayes

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
ypred1 = nb.predict(x_test)
```

```python
a = recall_score(ypred1,y_test)
b = accuracy_score(ypred1,y_test)
c = f1_score(ypred1,y_test)
d = precision_score(ypred1,y_test)

print(f"accuracy_score : {b} \n recall_score : {a} \n f1_score : {c} \n precision_score : {d}")
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm1 = confusion_matrix(ypred1,y_test)
cmd1 = ConfusionMatrixDisplay(cm1)
cmd1.plot()
plt.show()
```

```python
#K-means

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
ypred2 = knn.predict(x_test)
```

```python
a = recall_score(ypred2,y_test)
b = accuracy_score(ypred2,y_test)
c = f1_score(ypred2,y_test)
d = precision_score(ypred2,y_test)

print(f"accuracy_score : {b} \n recall_score : {a} \n f1_score : {c} \n precision_score : {d}")
```

```python
cm2 = confusion_matrix(ypred2,y_test)
cmd2 = ConfusionMatrixDisplay(cm2)
cmd2.plot()
plt.show()
```

```python
#SVM
from sklearn.svm import SVC

sv = SVC()
sv.fit(x_train,y_train)
```

```python
ypred3 = sv.predict(x_test)
```

```python
a = recall_score(ypred3,y_test)
b = accuracy_score(ypred3,y_test)
c = f1_score(ypred3,y_test)
d = precision_score(ypred3,y_test)

print(f"accuracy_score : {b} \n recall_score : {a} \n f1_score : {c} \n precision_score : {d}")
```

```python
cm3 = confusion_matrix(ypred3,y_test)
cmd3 = ConfusionMatrixDisplay(cm3)
cmd3.plot()
plt.show()
```
